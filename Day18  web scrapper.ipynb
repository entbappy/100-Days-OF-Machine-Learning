```python
company = soup.find_all('div', class_= 'company-content-wrapper')
```

```python
len(company)
```

```python
name = []
rating = []
reviews = []
ctype = []
location = []
how_old = []
no_of_employee = []

for com in company:
    name.append(com.find('h2').text.strip())
    rating.append(com.find('p','rating').text.strip())
    reviews.append(com.find('a' , class_='review-count').text.strip()[1:-1])
    ctype.append(com.find_all('p', class_='infoEntity')[0].text.strip())
    location.append(com.find_all('p', class_='infoEntity')[1].text.strip())
    how_old.append(com.find_all('p', class_='infoEntity')[2].text.strip())
    try:
        no_of_employee.append(com.find_all('p',class_= 'infoEntity')[3].text.strip())
    except:
        no_of_employee.append(np.nan)
        
```

```python
df = {'name':name,
   'rating':rating,
   'reviews':reviews,
   'company_type':ctype,
   'Head_Quarters':location,
   'Company_Age':how_old,
   'No_of_Employee':no_of_employee}

data = pd.DataFrame(df)
```

```python
final=pd.DataFrame()

for j in range(1,3):
    
    webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j)).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_='company-content-wrapper')
    
    name=[]
    rating=[]
    reviews=[]
    ctype=[]
    hq=[]
    how_old=[]
    no_of_employee=[]
    
    for i in company:
        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)

        try:
            rating.append(i.find('p',class_='rating').text.strip())
        except:
            rating.append(np.nan)

        try:
            reviews.append(i.find('a' , class_='review-count').text.strip())
        except:
            reviews.append(np.nan)

        try:
            ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)
            
        try:
            hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)

        try:
            how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except:
            how_old.append(np.nan)
            
        try:
            no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            no_of_employee.append(np.nan)

            
    df=pd.DataFrame({'name':name,
    'rating':rating,
    'reviews':reviews,
    'company_type':ctype,
    'Head_Quarters':hq,
    'Company_Age':how_old,
    'No_of_Employee':no_of_employee,
    })
  
    final=final.append(df,ignore_index=True)
```
